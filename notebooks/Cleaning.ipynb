```python
import pandas as pd
import numpy as np
```

```python
#Loading in dataset with 1 million rows since my computer can not handle all 6 million
df = pd.read_csv("../data/PS_20174392719_1491204439457_log.csv", nrows = 1000000)
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
#Generating one-hot-encoding for the payment type columns using pd.get_dummies

#Temporary one hot encodings with prefix type
Temp_OHE = pd.get_dummies(df["type"] , prefix = "type").astype(int)

#Final dataframe, joining one hot encodings
df_final = df.join(Temp_OHE)

df_final.drop("type" , axis = 1 , inplace = True)

df_final.head()
```

```python
df_final.shape
```

```python
df_final.isna().sum()
```

Since there are no missing values,  I can move on to performing any other neccesary transformations



```python
df_final.shape
```

```python
#Dropping the nameOrig and nameDest columns as they provide no information for the model
#Also dropping isFlaggedFraud because it incorrrectly flags all transactions over a certain amount
df_final = df_final.drop(["nameOrig" , "nameDest" , "isFlaggedFraud"] , axis = 1)

df_final.shape
```

```python
#Viewing the final dataframe before beginning the model training phase
df_final
```

```python
df_final.to_csv("../data/Fraud_Data_cleaned.csv",index=False)
```
